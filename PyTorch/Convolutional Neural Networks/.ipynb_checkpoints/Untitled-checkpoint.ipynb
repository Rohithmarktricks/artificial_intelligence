```python
def crop(image, cropx, cropy):
    y , x, c, = image.shape
    
    startx = x//2 - (cropx//8)
    starty = y//2 - (cropy//4)
    
    stopx = startx + cropx
    stopy = starty + 2*cropy
    
    return image[starty:stopy, startx:stopx]
```

```python
bird_cropped = crop(bird, 200, 200)
```

```python
plt.imshow(bird_cropped)
```

```python
# random noise

from skimage.util import random_noise

sigma = 0.155
noisy_bird = random_noise(bird, var = sigma**2)
```

```python
plt.imshow(noisy_bird)
```

```python
from skimage.restoration import denoise_tv_chambolle, estimate_sigma
```

```python
sigma_est = estimate_sigma(noisy_bird, multichannel=True, average_sigmas=True)
```

```python
sigma_est
```

```python
plt.imshow(denoise_tv_chambolle(noisy_bird, weight=0.1, multichannel=True))
```
