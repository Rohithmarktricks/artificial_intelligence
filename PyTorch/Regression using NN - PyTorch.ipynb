```python
import numpy as np
import torch
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_train = np.random.rand(25, 1, )
```

```python
y_train = np.random.rand(25, 1)
```

```python
x = torch.from_numpy(x_train)
y = torch.from_numpy(y_train)
```

```python
x= x.float()
y = y.float()
```

```python
inp = 1
hidden = 5
out = 1
```

```python
model = torch.nn.Sequential(
        torch.nn.Linear(inp, hidden),
        torch.nn.ReLU(),
        torch.nn.Linear(hidden, out))
```

```python
loss_fn = torch.nn.MSELoss(reduction='sum')
```

```python
learning_rate = 1e-4

for i in range(5000):
    y_pred = model(x)
    
    loss = loss_fn(y_pred, y)
    print(i, loss.item())
    
    model.zero_grad()
    loss.backward()
    
    with torch.no_grad():
        for param in model.parameters():
            param -= learning_rate * param.grad
```

```python
predicted_tensor = model(x)
```

```python
predicted_arr = predicted_tensor.detach().numpy()
```

```python
plt.figure(figsize=(8,8))

plt.scatter(x_train, y_train, c='green', s=250, label='Original data')

plt.plot(x_train, predicted_arr, label= 'Fitted line')

plt.legend()
plt.show()
```

### Using PyTorch optimizer

```python
learning_rate = 1e-6
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

for i in range(500):
    
    y_pred = model(x)
    
    loss = loss_fn(y_pred, y)
    
    print(i, loss.item())
    optimizer.zero_grad()
    
    loss.backward()
    
    optimizer.step()
    
```
